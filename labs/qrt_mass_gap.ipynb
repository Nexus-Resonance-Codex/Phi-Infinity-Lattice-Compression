### 1. Energy Spectrum Projection
We simulate a spectrum of quantum field fluctuations and test their stability against the 
$C_9$ manifold. We expect to see "Allowed" states separated by "Gaps" of chaotic instability.

```python
controller = LatticeResonanceController(target_dim=8192)

spectrum_range = np.linspace(0, 10, 1000)
stability_scores = []

for energy in spectrum_range:
    # Map energy to a lattice index
    root = controller._get_c9_cyclic_index(int(energy * 1000))
    # Score based on distance from chaotic singularites (3, 6, 9)
    score = 1.0 if root not in [3, 6, 9] else 0.0
    stability_scores.append(score)

plt.fill_between(
    spectrum_range, stability_scores, color='#00f0ff', 
    alpha=0.3, label='Allowed States'
)
plt.plot(spectrum_range, stability_scores, color='#00f0ff', linewidth=1)
plt.title('Quantum Field Stability: The Emergence of the Mass Gap')
plt.xlabel('Energy Potential (E)')
plt.ylabel('Manifold Stability (0=Gap, 1=Stable)')
plt.legend()
plt.show()
```

### 2. Resonance Harmonic Distribution
Visualizing the quantized "allowed" energy levels as harmonic peaks. The empty space 
between peaks is the functional Mass Gap.

```python
frequencies = np.linspace(0.1, 5, 500)
resonance = []

for f in frequencies:
    # Formula representing QRT harmonic tension
    val = np.abs(np.cos(f * np.pi * 1.618))
    # Apply C9 filter
    if controller._get_c9_cyclic_index(int(val * 10000)) in [3, 6, 9]:
        resonance.append(0)
    else:
        resonance.append(val)

plt.plot(frequencies, resonance, color='#ff3366', linewidth=2)
plt.fill_between(frequencies, resonance, color='#ff3366', alpha=0.1)
plt.title('Harmonic Mass Quantization via $\\varphi^\\infty$ Resonance')
plt.xlabel('Field Vibration Frequency')
plt.ylabel('Quantized Probability Density')
plt.show()
```
