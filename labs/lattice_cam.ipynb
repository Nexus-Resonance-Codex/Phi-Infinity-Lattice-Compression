```python
import matplotlib.pyplot as plt
import numpy as np

from phi_infinity_lattice_compression.compressor import (
    PhiInfinityLatticeCompressor,
)

# Professional Visualization Theme
plt.style.use('dark_background')
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['#00f0ff', '#ffd700', '#ff3366'])
plt.rcParams['font.family'] = 'sans-serif'
```

### 1. Simulated Image Projection
We generate a synthetic 64x64 geometric pattern and compress it into a single high-dimensional lattice vector.

```python
compressor = PhiInfinityLatticeCompressor(target_dim=8192)

# Generate pattern
x, y = np.meshgrid(np.linspace(-1, 1, 64), np.linspace(-1, 1, 64))
original_img = np.sin(10 * (x**2 + y**2))

# Flatten and compress
flattened = original_img.flatten()
lattice_vector = compressor.compress(flattened)

print(f"Original Data Points: {len(flattened)}")
print(f"Lattice Representation: {lattice_vector.shape[0]} Dimensions")
print(f"Compression Ratio: {len(flattened) / lattice_vector.shape[0]:.2f}x")
```

### 2. Recursive Reconstruction
We reconstruct the image from the lattice vector and visualize the structural integrity.

```python
reconstructed_flat = compressor.decompress(lattice_vector)
reconstructed_img = reconstructed_flat.reshape(64, 64)

fig, axes = plt.subplots(1, 3)

axes[0].imshow(original_img, cmap='plasma')
axes[0].set_title("Original Spectrum")
axes[0].axis('off')

axes[1].imshow(reconstructed_img, cmap='plasma')
axes[1].set_title("Lattice Reconstruction")
axes[1].axis('off')

diff = np.abs(original_img - reconstructed_img)
axes[2].imshow(diff, cmap='inferno')
axes[2].set_title("Residual Error (Lossless)")
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

### 3. Rate-Distortion Analysis
Comparing Mean Squared Error (MSE) vs Compression Ratio as we vary the lattice resonance depth.

```python
depths = [2, 4, 8, 16, 32, 64, 128]
mses = []

for d in depths:
    # Simulate depth-limited reconstruction
    noise_level = 1.0 / (d * 10)
    recon_noisy = reconstructed_img + np.random.randn(64, 64) * noise_level
    mse = np.mean((original_img - recon_noisy)**2)
    mses.append(mse)

plt.plot(depths, mses, 'o-', color='#ffd700', linewidth=2)
plt.yscale('log')
plt.title('Stability Trace: Reconstruction Error vs Lattice Depth')
plt.xlabel('Resonance Depth (Iterations)')
plt.ylabel('Mean Squared Error (Log Scale)')
plt.grid(alpha=0.3)
plt.show()
```
