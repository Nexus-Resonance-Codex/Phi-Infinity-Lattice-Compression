### 1. Simulating Quantum Decoherence
We model a quantum state vector over a series of gate operations. As the circuit deepens, 
standard operations accumulate "noise" (stochastic drift).

```python
controller = LatticeResonanceController(target_dim=8192)

gate_steps = 100
coherence_standard = []
coherence_stabilized = []

state_standard = np.random.randn(8192).astype(np.float64)
state_standard /= np.linalg.norm(state_standard)

state_stabilized = state_standard.copy()

for _i in range(gate_steps):
    # Standard accumulation of gate noise
    noise = np.random.randn(8192) * 0.05
    state_standard = state_standard + noise
    state_standard /= np.linalg.norm(state_standard)

    # Stabilized accumulation via Lattice Resonance Modulation
    state_stabilized = state_stabilized + noise
    state_stabilized = controller.modulate_resonance(state_stabilized)
    state_stabilized /= np.linalg.norm(state_stabilized)

    # Measure Coherence (Projection onto ideal manifold)
    coherence_standard.append(np.abs(np.vdot(state_standard, state_standard)))
    coherence_stabilized.append(np.abs(np.vdot(state_stabilized, state_stabilized)))

plt.plot(coherence_standard, "--", label="Standard Decoherence", alpha=0.6)
plt.plot(coherence_stabilized, label="Resonance-Stabilized Coherence", linewidth=3)
plt.title("Quantum State Stability: Resonance Suppression vs Noise Accumulation")
plt.xlabel("Circuit Gate Depth")
plt.ylabel("Coherence Fidelity")
plt.legend()
plt.show()
```
