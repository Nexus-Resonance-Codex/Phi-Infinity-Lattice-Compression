### 1. Resonance Embedding & Conformational Distance
We compare the wild-type sequence against a single-point mutation and a completely distinct protein fold purely through lattice topology.

```python
accelerator = ProteinLatticeAccelerator()

# Sequence A (Wild Type - Hemoglobin-like)
seq_a = (
    "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSHGSAQVKGHGKKVADALTNAVAHVDDMP"
    "NALSALSDLHAHKLRVDPVNFKLLSHCLLVTLAAHLPAEFTPAVHASLDKFLASVSTVLTSKYR"
)

# Sequence B (Mutation - Similar structure)
seq_b = (
    "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSHGSAQVKGHGKKVADALTNAVAHVDDMP"
    "NALSALSDLHAHKLRVDPVNFKLLSHCLLVTLAARLPAEFTPAVHASLDKFLASVSTVLTSKYR"
)

# Sequence C (Distinct Fold - Ras-like)
seq_c = (
    "MTEYKLVVVGAGGVGKSALTIQLIQNHFVDEYDPTIEDSYRKQVVIDGETCLLDILDTAGQEEYSAMRDQYMRTGEGF"
    "LCVFAINNTKSFEDIHQYREQIKRVKDSDDVPMVLVGNKCDLAARTVESRQAQDLARSYGIPYIE"
)

dist_ab = accelerator.conformational_distance(seq_a, seq_b)
dist_ac = accelerator.conformational_distance(seq_a, seq_c)

print("Lattice Conformational Distance:")
print(f"  [A vs B (Mutation)]:  {dist_ab:.8f} (Low Tension)")
print(f"  [A vs C (Distinct)]:  {dist_ac:.8f} (High Dispersion)")
```

### 2. Topological Tension Resonance Map
This visualization projects the 8192D embedding residues into a flattened radial space using $\varphi$ angular steps. Clusters indicate strong local secondary structure resonance.

```python
vec = accelerator.embed_protein(seq_a)
active = np.where(vec != 0)[0][:300]
mags = vec[active]

phi = (1 + 5**0.5) / 2
theta = active * (360.0 / phi**2)
r = np.abs(mags)

fig, ax = plt.subplots(subplot_kw={"projection": "polar"})
sc = ax.scatter(np.deg2rad(theta), r, c=r, cmap="plasma", s=60, edgecolors="none", alpha=0.8)
ax.grid(color="#333", linewidth=0.5)
ax.set_facecolor("#000000")
plt.title("Bio-Lattice Resonance Map (Sequence A)", fontdict={"color": "#FFD700", "size": 14})
plt.colorbar(sc, label="Resonance Amplitude")
plt.show()
```
