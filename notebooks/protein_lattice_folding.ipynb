```python
import matplotlib.pyplot as plt
import numpy as np

from phi_infinity_lattice_compression.protein_accelerator import (
    ProteinLatticeAccelerator,
)

plt.style.use("dark_background")
plt.rcParams["figure.figsize"] = (10, 6)
```

### Projecting Sequences into the Golden Lattice
We compare homology distance between sequences without requiring structural simulation, 
purely through the geometric properties of the projection.

```python
accelerator = ProteinLatticeAccelerator()

# Sequence A (Wild Type)
seq_a = (
    "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSHGSAQVKGHGKKVADALTNAVAHVDDMPN"
    "ALSALSDLHAHKLRVDPVNFKLLSHCLLVTLAAHLPAEFTPAVHASLDKFLASVSTVLTSKYR"
)

# Sequence B (Similar structure, small mutation)
seq_b = (
    "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHFDLSHGSAQVKGHGKKVADALTNAVAHVDDMPN"
    "ALSALSDLHAHKLRVDPVNFKLLSHCLLVTLAARLPAEFTPAVHASLDKFLASVSTVLTSKYR"
)

# Sequence C (Completely distinct)
seq_c = (
    "MTEYKLVVVGAGGVGKSALTIQLIQNHFVDEYDPTIEDSYRKQVVIDGETCLLDILDTAGQEEYSAMRDQYMRTGEGFL"
    "CVFAINNTKSFEDIHQYREQIKRVKDSDDVPMVLVGNKCDLAARTVESRQAQDLARSYGIPYIE"
)

dist_ab = accelerator.conformational_distance(seq_a, seq_b)
dist_ac = accelerator.conformational_distance(seq_a, seq_c)

print(f"Structural Difference A vs B (Mutation): {dist_ab:.6f}")
print(f"Structural Difference A vs C (Unrelated): {dist_ac:.6f}")
```

### Phase State Visualization
A 2D radial plot of the resonant boundaries for a single sequence embedding.

```python
vec = accelerator.embed_protein(seq_a)
active_indices = np.where(vec != 0)[0][:200]  # First 200 resonant fields
magnitudes = vec[active_indices]

# Generate polar coordinates mimicking the 8192D spiral flattening
theta = active_indices * (360.0 / (1.6180339887**2))
r = np.abs(magnitudes)

fig, ax = plt.subplots(subplot_kw={"projection": "polar"})
ax.scatter(theta, r, c=r, cmap="plasma", s=50, alpha=0.7)
ax.grid(color="#333")
ax.set_title("Topological Tension Resonance Map", fontdict={"color": "#FFD700"})
plt.show()
```
