# φ^∞ Lattice Compression: Infinite Context Engine Demonstration

Under the **Trageser Tensor Theorem (TTT)**, standard LLM context windows fail exponentially due to their $O(N^2)$ algorithmic constraints via raw multi-head attention. 
This notebook demonstrates the mathematical substitution of the KV cache with a **Golden-Ratio (φ) damped Residual Hierarchy**, fundamentally achieving $O(1)$ memory scaling boundaries.

```python
import matplotlib.pyplot as plt
import numpy as np

from phi_infinity_lattice_compression.residual_hierarchy import (
    QRTDampedResidualHierarchy,
)

# Set TTT visualizer theme for plotting
plt.style.use("dark_background")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.prop_cycle"] = plt.cycler(color=["#00f0ff", "#ffd700", "#ff3366"])
```

### Simulating Context Absorption
We will feed 5,000 contextual embeddings (tokens) into the `QRTDampedResidualHierarchy` and extract the exponential hardware memory savings.

```python
hierarchy = QRTDampedResidualHierarchy(target_dim=8192)

simulated_token_count = 5000
memory_tracking_standard = []
memory_tracking_phi = []

for _i in range(simulated_token_count):
    # Simulating a 4096D internal projection from a model like Llama-3
    dummy_state = np.random.randn(8192)
    hierarchy.add_context(dummy_state)

    stats = hierarchy.get_memory_usage()
    memory_tracking_standard.append(stats["raw_memory_mb"])
    memory_tracking_phi.append(stats["hierarchy_memory_mb"])

print(f"Final Classical KV Cache Size: {memory_tracking_standard[-1]:.2f} MB")
print(f"Final φ^∞ Lattice Size: {memory_tracking_phi[-1]:.2f} MB")
```

### Visualizing O(1) Bounding vs O(N) Linear Growth
Comparing the physical VRAM footprint.

```python
plt.plot(memory_tracking_standard, label="Classical KV Cache O(N)")
plt.plot(memory_tracking_phi, label="φ^∞ Hierarchy O(1)")
plt.title("Memory Scaling: Context vs Dimension")
plt.xlabel("Tokens Processed")
plt.ylabel("Memory (MB)")
plt.legend()
plt.tight_layout()
plt.show()
```
